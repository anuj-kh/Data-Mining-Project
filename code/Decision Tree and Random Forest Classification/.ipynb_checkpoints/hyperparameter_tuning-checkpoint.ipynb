# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Defining months

```python
month=["April 2018","May 2018","June 2018","July 2018","August 2018","September 2018","October 2018","November 2018","December 2018","January 2019","February 2019","March 2019"]
```

# Averaging over 12 months data for Max depth vs. Accuracy in Decision tree

```python
df=[]
for i in month:                                              # appending dataframe of all the months to df
    df.append(pd.read_csv("./result/hyperparameter tuning/max depth/maxDepth_{}.csv".format(i)))   
```

```python
# merging all the dataframes with respect to max depth, in order to find average accuracy
df2=pd.merge(df[0], df[1], on='Max Depth')
df2=pd.merge(df2, df[2], on='Max Depth')
for i in range(12):
    if i<=2:
        continue
    df2=pd.merge(df2, df[i], on='Max Depth')
```

```python
df2.head()
```

```python
df2=df2.drop('Max Depth', axis = 1)                           # dropping max depth
mean=df2.mean(axis=1)                                         # and finding mean row wise, i.e., for each max depth value
```

```python
df[0]["Accuracy"]=mean                                        # storing that into a dataframe
print(df[0][df[0]['Accuracy']==df[0]['Accuracy'].max()])      # printing max. accuracy among average accuracies
```

```python
# storing that into csv file
df[0].to_csv('./result/hyperparameter tuning/max depth/maxDepth_average(12 months).csv',index=False,header=True)
```

# Averaging over 12 months data for No. of trees vs. Accuracy in Random forest

```python
df=[]
for i in month:                                                 # appending dataframe of all the months to df
    df.append(pd.read_csv("./result/hyperparameter tuning/no of trees/noOfTrees_{}.csv".format(i)))
```

```python
# merging all the dataframes with respect to no. of trees, in order to find average accuracy
df2=pd.merge(df[0], df[1], on='No. of trees')
df2=pd.merge(df2, df[2], on='No. of trees')
for i in range(12):
    if i<=2:
        continue
    df2=pd.merge(df2, df[i], on='No. of trees')
```

```python
df2.head()
```

```python
df2=df2.drop('No. of trees', axis = 1)                       # dropping no. of trees
mean=df2.mean(axis=1)                                        # and finding mean row wise, i.e., for each no. of trees value
```

```python
df[0]["Accuracy"]=mean                                       # storing that into a dataframe
print(df[0][df[0]['Accuracy']==df[0]['Accuracy'].max()])     # printing max. accuracy among average accuracies
```

```python
# storing that into csv file
df[0].to_csv('./result/hyperparameter tuning/no of trees/noOfTrees_average(12 months).csv',index=False,header=True)
```

# Plotting for Max depth vs. Accuracy in Decision tree

```python
# plotting max depth vs. accuracy

month=["April 2018","May 2018","June 2018","July 2018","August 2018","September 2018","October 2018","November 2018","December 2018","January 2019","February 2019","March 2019","average(12 months)","combined_data(12 months)"]
for i in month:
    df=pd.read_csv("./result/hyperparameter tuning/max depth/maxDepth_{}.csv".format(i))
    plt.plot(df["Max Depth"],df["Accuracy"])
    plt.xlabel("Max Depth")
    plt.ylabel("Accuracy")
    plt.title(i)
    plt.savefig("../../plots/Decision Tree and Random Forest Classification/max depth/{}.png".format(i))
    plt.show()
```

```python
# finding max. average accuracy
df=pd.read_csv("./result/hyperparameter tuning/max depth/maxDepth_{}.csv".format(month[12]))
print("Max average accuracy over 12 months: \n",df[df['Accuracy']==df['Accuracy'].max()])
```

```python
# finding max. total accuracy
df2=pd.read_csv("./result/hyperparameter tuning/max depth/maxDepth_{}.csv".format(month[13]))
print("\nMax total accuracy over 12 months: \n",df2[df2['Accuracy']==df2['Accuracy'].max()])
```

# Plotting for No. of trees in Random forest vs. Accuracy

```python
# plotting no. of trees vs. accuracy

month=["April 2018","May 2018","June 2018","July 2018","August 2018","September 2018","October 2018","November 2018","December 2018","January 2019","February 2019","March 2019","average(12 months)"]
for i in month:
    df=pd.read_csv("./result/hyperparameter tuning/no of trees/noOfTrees_{}.csv".format(i))
    plt.plot(df["No. of trees"],df["Accuracy"])
    plt.xlabel("No. of trees")
    plt.ylabel("Accuracy")
    plt.title(i)
    plt.savefig("../../plots/Decision Tree and Random Forest Classification/no of trees/{}.png".format(i))
    plt.show()
```

```python
# finding max. average accuracy
df=pd.read_csv("./result/hyperparameter tuning/no of trees/noOfTrees_{}.csv".format(month[12]))
print("Max average accuracy over 12 months: \n",df[df['Accuracy']==df['Accuracy'].max()])
```
