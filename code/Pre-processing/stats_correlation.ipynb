# Correlation functions:

```python
def result(coeff):
    coeff2=abs(coeff)
    if coeff2>0 and coeff2<0.2:
        print("Very Weak", end =" ")
    elif coeff2>=0.2 and coeff2<0.4:
        print("Weak", end =" ")
    elif coeff2>=0.4 and coeff2<0.6:
        print("Moderate", end =" ")
    elif coeff2>=0.6 and coeff2<0.8:
        print("Strong", end =" ")
    elif coeff2>=0.8 and coeff2<1.0:
        print("Very strong", end =" ")
    elif coeff2==1.0:
        print("Perfect", end =" ")
    else:
        print("No correlation !!")
    if coeff<0:
        print("negatively correlated !!")
    elif coeff>0:
        print("positively correlated !!")
        
def pearson_cor(X,Y):
    X=X.astype(float)
    Y=Y.astype(float)
    covariance=np.cov(X,Y)
    #np.cov calculates covariance
    pearson=covariance/(np.std(X)*np.std(Y))
    #pearson's correlation =(covariance of X and Y)/(standard deviation of X * standard deviation of Y)
    coeff=pearson[0,1]
    """
    pearson will hold the correlation matrix, 
    and the correlation coefficient will the element at 0,1 and 1,0. 
    So multiply with either [0,1] or [1,0]
    """
    print("\033[1m" + "Pearson's:"+"\033[0m")
    print("correlation matrix:\n",pearson)
    print("correlation coefficient=",coeff)
    result(coeff)
    print("\n")
    
def spearman_cor(X,Y):
    X=X.astype(float)
    Y=Y.astype(float)
    #.astype(float) type casts value to floating data type
    
    rankX=pd.Series(X).rank()
    rankY=pd.Series(Y).rank()
    #pd.Series(X).rank() calculates rank
    
    covariance2=np.cov(rankX,rankY)
    #np.cov calculates covariance
    spearman=covariance2/(np.std(rankX)*np.std(rankY))
    #pearson's correlation =(covariance of X and Y)/(standard deviation of X * standard deviation of Y)
    coeff=spearman[0,1]
    """
    spearman will hold the correlation matrix, 
    and the correlation coefficient will the element at 0,1 and 1,0. 
    So multiply with either [0,1] or [1,0]
    """
    print("\033[1m" + "Spearman's:"+"\033[0m")
    print("correlation matrix:\n",spearman)
    print("correlation coefficient=",coeff)
    result(coeff)
    print("\n")
```
