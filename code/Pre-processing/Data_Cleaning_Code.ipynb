#### STORING MONTHLY DATA ADDRESSES IN A LIST 

The addresses of monthly data files are fed into a list which will facilitate the loop in the next code block.

```python
#storing the addresses of the monthly data sets
addresses = [r"../../data/May 2018/May_2018.csv"
           ,r"../../data/November 2018/November_2018.csv"
           ,r"../../data/October 2018/October_2018.csv"
           ,r"../../data/September 2018/September_2018.csv"
           ,r"../../data/April 2018/April_2018.csv"
           ,r"../../data/August 2018/August_2018.csv"
           ,r"../../data/December 2018/December_2018.csv"
           ,r"../../data/February 2019/February_2019.csv"
           ,r"../../data/January 2019/January_2019.csv"
           ,r"../../data/July 2018/July_2018.csv"
           ,r"../../data/June 2018/June_2018.csv"
           ,r"../../data/March 2019/March_2019.csv"
           ]
```

#### LOOP TO FIND STATES FROM GIVEN COORDINATES 

The loop runs through monthly data to check for entries with coordinates given and missing state names, which are then processed through the reverse geocoder to output the previously unknown state names

```python
#looping through the addresses
for j in addresses:
    file_name=j
    
    #opening the .csv file
    df = pd.read_csv(file_name)
    
    #storing the directary path in 'path' variable
    path=os.path.dirname(file_name)
    
    #storing the base name in the 'file' variable
    file=os.path.basename(file_name)
    
    #print(path)
    #print(file)
    
    #Editing the dataframe as per the following coditions:
    #network type should be known
    df = df[df['Network Type']!='Unknown']
    
    #Position cooordinates cannot be -1 
    df = df[df.Latitude != -1]
    df = df[df.Longitude != -1]
    
    #Either the state name is valid or if we it is invalid, we have valid location coordinates 
    df=df[((df['State Name'].isna()) & (df.Latitude!=-1) & (df.Longitude!=-1))|(df['State Name'].notna())]
    
    #resetting the index of the dataframe so that we get idices as consecutive whole numbers
    df=df.reset_index(drop=True)

#block to find out the state of the objects which have valid location coordinates but not the state name
    
    # Driver function 
    if __name__=="__main__": 
        for i in df.index:
            #if we already have the state name than no need to process the current object
            if(df['State Name'].notna()[i]):
                continue
            
            #storing the longitude and latitude in the respective variables and then making a tuple
            long=df['Longitude'][i]
            lat=df['Latitude'][i]
            coordinates =(lat, long)
            
            #getting the address of the location
            result = rg.search(coordinates) 
            print(i,result)

            #if country code is not 'IN' ie country is not India then set 'state name' as 'NA'
            if(result[0]['cc']!='IN'):
                df.at[i,'State Name']= 'NA'
                #print(i,lat,long,"NA Given")
                continue
            #set the 'state name' as the state that is obtained from the address    
            df.at[i,'State Name']= result[0]['admin1']
        

    
```
